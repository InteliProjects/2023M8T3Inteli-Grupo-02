```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Documentação do teste automatizado da simulação de Markov

Esse notebook irá documentar e analisar os resultados obtidos no teste automatizado do sistema novo, presente nesta mesma pasta com nome novo.test.js

O teste simula uma cadeia de Markov para duas máquinas diferentes em quatro cenários distintos: 10, 20, 30 e 40 passos. Cada teste define uma matriz de transição que representa as probabilidades de mudança de estado para cada máquina. Para executar esse código é necessário inserir o comando ``npm run test`` no root deste diretório, que ele irá executar todas as suites de testes.

Após executar o teste, os resultados são salvos em arquivos JSON, onde nesse notebook serão utilizados para construção de gráficos. 

## Visualização dos resultados

Para visualização dos resultados é necessário ter executado o teste conforme descrito na documentação acima.

```python
with open('testResults10Steps.json') as file:
    data = json.load(file)


df = pd.DataFrame(data)

df['Estado Máquina 1'] = df['statesMachine1'].map({0: 'Funcionando', 1: 'Não Funcionando'})
df['Estado Máquina 2'] = df['statesMachine2'].map({2: 'Funcionando', 3: 'Não Funcionando'})


df['Passo'] = df.index
bar_width = 0.35
index = np.arange(len(df))


fig, ax = plt.subplots()


bars1 = ax.bar(index - bar_width/2, df['Estado Máquina 1'] == 'Funcionando', bar_width, label='Máquina 1 - Funcionando', color='green')
bars2 = ax.bar(index - bar_width/2, df['Estado Máquina 1'] == 'Não Funcionando', bar_width, label='Máquina 1 - Não Funcionando', color='red', bottom=df['Estado Máquina 1'] == 'Funcionando')


bars3 = ax.bar(index + bar_width/2, df['Estado Máquina 2'] == 'Funcionando', bar_width, label='Máquina 2 - Funcionando', color='green')
bars4 = ax.bar(index + bar_width/2, df['Estado Máquina 2'] == 'Não Funcionando', bar_width, label='Máquina 2 - Não Funcionando', color='red', bottom=df['Estado Máquina 2'] == 'Funcionando')


ax.set_xlabel('Passos')
ax.set_ylabel('Estado')
ax.set_title('Estados das Máquinas em Cada Passo')
ax.set_xticks(index)
ax.set_xticklabels(df['Passo'])
ax.legend()


plt.show()
```
