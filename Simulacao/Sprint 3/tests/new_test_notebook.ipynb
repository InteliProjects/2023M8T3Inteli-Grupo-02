```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('testResults10Steps.json') as file:
    data = json.load(file)

# Converter dados para DataFrame
df = pd.DataFrame(data)

# Plotar gráfico de linha para as máquinas
plt.plot(df['statesMachine1'], label='Máquina 1')
plt.plot(df['statesMachine2'], label='Máquina 2')
plt.xlabel('Passos')
plt.ylabel('Estado')
plt.title('Estados das Máquinas ao Longo do Tempo')
plt.legend()
plt.show()
```
